In order to visualize the coorelation between upregulated and downregulated Differentially Expressed Tif3 genes after 3% urea treatment and without treatment in both wild and mutant strains. 

#### Importing Libraries

```python
import numpy as np  # In regard to perform mathematical operations on array
import matplotlib.pyplot as plt #Matplotlib is a plotting library 
# Pyplot is a state-based interface to a Matplotlib module to provide a MATLAB-like interface 
```

#### Plotting Bar Plot 

```python
N=6  #The total number of bars will be 6 as there were 6 types of comparison in Differential Gene Expression 
Upregulated_Genes=(54,8,65,63,23,13) 
Downregulated_Genes=(82,12,175,141,87,32)
ind=np.arange(N)  #The arange() function is used to get evenly spaced values within a given interval
width=0.35 #The width of each bar is determined 0.35

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, Upregulated_Genes, width,color="orange") #plotting bars for upregulated genes which will be at the top
p2 = plt.bar(ind, Downregulated_Genes, width,
             bottom = Upregulated_Genes,color="grey") #plotting bars for downregulated genes which will be at the bottom 
 
plt.title('Upregulated and Downregulated Genes')  #Giving title to the bar plot
#Bar names at the x-axis
plt.xticks(ind, ('tif3 wild type untreated vs 3% urea treated', 'tif3[delta]ntd untreated vs 3% urea treated', 'tif3 wild type vs tif3[delta]ntd untreated', 'tif3 wild 3% urea treated vs tif3[delta]ntd untreated', 'tif3 wild untreated vs tif3[delta]ntd 3% urea treated','tif3 wild vs tif3[delta]ntd 3% urea treated'),rotation=92)
plt.yticks(np.arange(0, 250, 10)) #y axis scaling from 0 to 250 with 10 intervals 
plt.legend((p1[0], p2[0]), ('Upregulated Genes', 'Downregulated Genes')) #plotting legends name
#plt.grid(zorder=52)

for r1, r2 in zip(p1, p2):   # zip() accepted p1 qnd p2 for upregulated and downregulated genes and merged them into a single tuple
    h1 = r1.get_height() #here, get_height return the height of the bar charts of upregulated genes (p1)
    h2 = r2.get_height() #here, get_height return the height of the bar charts of downregulated genes (p2)
#plt.text use to add text on plot. ha and va stands for horizontal and vertical allignment. r1.get_x()+r1.get_width()/2 or r2.get_x()+r2.get_width()/2 use for annotation in the x-coordinate 
# 'color'detect the text color,fontweight detects the thickness or tinness of the font in plot, fontsize detect the size of the font     
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1, ha="center", va="bottom", color="white", fontsize=7, fontweight="bold")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2, ha="center", va="bottom", color="white", fontsize=7, fontweight="bold")

# in regard to visualize the bar chart    
plt.show()

```
